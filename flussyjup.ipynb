```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib.dates
import re
import csv
import datetime
```

```python
for j in range(1,10):
    code_column = []
    numeric_column = []

    # Leggi i dati dal file
    with open(f'/home/admin/Documents/Dottorato Dalla/dati comune spire/flussi/FLUSSI2023050{j}.txt', 'r') as file:
    # Leggi ogni riga dal file
        lines = file.readlines()
        i = 0
        while i < len(lines):
            # Controlla se la riga inizia con "Section" e ha almeno 5 elementi (la stringa "Section" e quattro numeri)
            if lines[i].startswith('Section') and len(lines[i].split()) >= 5:
            # Estrai la stringa "Section" seguita dai primi quattro numeri
                section_info = lines[i].strip()
            # Estrai i 288 valori numerici dalle due righe successive e separali
                numeric_values = list(map(int, lines[i + 1].strip().split()))
                numeric_values.extend(list(map(int, lines[i + 2].strip().split())))
            # Aggiungi i dati alle rispettive colonne
                code_column.append(section_info)
                numeric_column.append(numeric_values)
            # Vai alla riga successiva alla fine delle tre righe elaborate
                i += 3
            else:
            # Se la riga non contiene i dati attesi, vai alla riga successiva
                i += 1

    merged_data = list(zip(code_column, numeric_column))
    # Scrivi i dati nel file di output

    # Scrivi i dati nel file CSV
    with open(f'FLUSSI2023050{j}_P.csv', 'w', newline='') as csvfile:
        # Crea un oggetto writer del CSV
        csvwriter = csv.writer(csvfile)
    
        # Scrivi l'intestazione del CSV, se necessario
        csvwriter.writerow(['Section', 'Numeric Data'])  # Modifica gli header secondo le tue esigenze
    
        # Scrivi i dati nel CSV
        csvwriter.writerows(merged_data)

for k in range(10, 32):
    code_column = []
    numeric_column = []

    with open(f'/home/admin/Documents/Dottorato Dalla/dati comune spire/flussi/FLUSSI202305{k}.txt', 'r') as file:
        lines = file.readlines()
        i = 0
        while i < len(lines):
            
            if lines[i].startswith('Section') and len(lines[i].split()) >= 5:
                section_info = lines[i].strip()
                numeric_values = list(map(int, lines[i + 1].strip().split()))
                numeric_values.extend(list(map(int, lines[i + 2].strip().split())))
                code_column.append(section_info)
                numeric_column.append(numeric_values)
                i += 3
            else:
                i += 1

    merged_data = list(zip(code_column, numeric_column))

    with open(f'FLUSSI202305{k}_P.csv', 'w', newline='') as csvfile:
        csvwriter = csv.writer(csvfile)
    
        csvwriter.writerow(['Section', 'Numeric Data'])  
    
        csvwriter.writerows(merged_data)
```
